### Outlier Treatment

```python
for fea in data.keys():
    
    Q1 = np.percentile(data[fea], q=25)
    
    Q3 = np.percentile(data[fea], q=75)
    
    iqr = Q3 - Q1
    step = (iqr * 1.5)
    
    print("Outliers for '{}':".format(fea))
    display( data[~((data[fea] >= Q1 - step) & (data[fea] <= Q3 + step))] )
    
```

** There are some outliers present in the data. But , I am not going to remove them because it contains important data like waht is the cost of a person who smokes and had 4 children, it is a rare case but happens if we remove outliers we will miss that data so, i'm not going to remove the outliers.**

Adaptive Boost Regressor algorithm is less affected to Over-Fitting because it will allocate weights the some set of the weak learners and then the algorithem give them to strong learner and as it is an iterative process, it continues to add learners until a limit is reached in the number of models.

## Implementation

### Performance of Benchmark model

```python
model = linear_model.LinearRegression()

model.fit(X_train, y_train)     

y_preds = model.predict(X_test)

print('r2_score is{:.2f}'.format(r2_score( y_test , y_preds)))
```

** From above we can say that , the r2_ score of benchmark model is low ( around 71 % ). **

### Testing Different Algorithms

```python
clf1 = DTR()
clf2 = ada()
clf3=RFR()
lis=[clf1,clf2,clf3]
for i in lis:
    i.fit(X_train, y_train)     

    y_preds = i.predict(X_test)
    print (i)
    print('r2_score is{:.2f}'.format(r2_score( y_test , y_preds)))
    print()
```

**By using all three algorithms we got good r2_score( around 82%) for Random forest Regressor so i choose Random forest Regressor as my Final model and i try to optimise and increase the perfomance of our model.**

### Function to Calculate r2_score

```python
def per_metric(y_true, y_predict):
    score =r2_score(y_true, y_predict)
    # Return the score
    return (score)
```

## Refinement

### Hyper Parameter Tuning using Grid Search

```python
    regressor = RFR()
    scoring_fnc = make_scorer(per_metric)
    param_grid = {'n_estimators': [90,130],
                   'max_depth': [4,5],
                   'min_samples_leaf':[1,5],
                   'min_samples_split' :[3,5],                  
                }
    grid = GridSearchCV(estimator=regressor,param_grid=param_grid,scoring = scoring_fnc)
    grids = grid.fit(X_train, y_train)

    # Return the optimal model after fitting the data
    best=grids.best_estimator_
    print(grids.best_estimator_)
```

### Checking performance of the Optimized model

```python
best.fit(X_train , y_train)     # Calculating the training time
y_preds = best.predict(X_test)
grid = GridSearchCV(estimator=regressor,param_grid=param_grid,scoring = scoring_fnc)
best=grids.best_estimator_    
print('r2_score is :{:.2f}'.format(r2_score(y_test, y_preds)))
```

** So, after optimising the Random forest Regressor algorithm the performance of the model is increased (r2_score changed from 82% to 84%) which is good.**
